```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import ImageTk, Image

def checkIfTrue(im2):
    if im2[picture_size//2][picture_size//2] > 0.99:
        return True
    else: 
        return False
    
picture_size = 20

PATH = os.getcwd()
data_path1 = PATH + '\\foto2\\wejsciowe'
data_path2 = PATH + '\\foto2\\wyjsciowe'

#data_dir_list = os.listdir(data_path)
img_list = []

for img1, img2 in zip(os.listdir(data_path1), os.listdir(data_path2)):
    licznikTrue = 0
    licznikFalse = 0
    input_img1 = Image.open(data_path1 + '\\' + img1)
    input_img2 = Image.open(data_path2 + '\\' + img2)
    print(img1,img2)
    input_img1 = np.array(input_img1)
    input_img2 = np.array(input_img2)
    #print(input_img1.shape[0], input_img2.shape[1])
    for r in range(0,input_img1.shape[0] - picture_size, picture_size):
        for c in range(0,input_img1.shape[1] - picture_size, picture_size):
            small_input_img1 = input_img1[r:r+picture_size,c:c+picture_size]
            small_input_img2 = input_img2[r:r+picture_size,c:c+picture_size]
            if checkIfTrue(small_input_img2):
                img_list.append(tuple((small_input_img1,"TRUE")))
                licznikTrue += 1
            else:
                img_list.append(tuple((small_input_img1,"FALSE")))
                licznikFalse += 1

    #print(img_list)
    print(licznikTrue,"/", licznikFalse) 

    
```
