```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from PIL import ImageTk, Image

def checkIfTrue(im2):
    if im2[picture_size//2][picture_size//2] > 0.99:
        return True
    else: 
        return False
    
picture_size = 20

PATH = os.getcwd()
data_path1 = PATH + '\\foto2\\wejsciowe'
data_path2 = PATH + '\\foto2\\wyjsciowe'

#data_dir_list = os.listdir(data_path)
samples=[]
targets=[]

for img1, img2 in zip(os.listdir(data_path1), os.listdir(data_path2)):
    licznikTrue = 0
    licznikFalse = 0
    input_img1 = Image.open(data_path1 + '\\' + img1)
    input_img2 = Image.open(data_path2 + '\\' + img2)
    print(img1,img2)
    input_img1 = np.array(input_img1)
    input_img2 = np.array(input_img2)
    #print(input_img1.shape[0], input_img2.shape[1])
    for r in range(0,input_img1.shape[0] - picture_size, picture_size):
        for c in range(0,input_img1.shape[1] - picture_size, picture_size):
            small_input_img1 = input_img1[r:r+picture_size,c:c+picture_size]
            small_input_img2 = input_img2[r:r+picture_size,c:c+picture_size]
            
            if checkIfTrue(small_input_img2):
                samples.append(small_input_img1.flatten())
                targets.append(1)
                licznikTrue += 1
            else:
                if licznikFalse<=licznikTrue:
                    samples.append(small_input_img1.flatten())
                    targets.append(0)
                    licznikFalse += 1
    #print(img_list)
    #print(licznikTrue,"/", licznikFalse) 


#Poki co 10000, aby nie przepełnić pamięci
learn_size=10000
X=samples[:learn_size]
y=targets[:learn_size]

test_size=1000
Xtest=samples[learn_size:learn_size+test_size]
ytest=targets[learn_size:learn_size+test_size]

clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
#clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X, y)

correct_results=0
for idx,test in enumerate(Xtest):
    response = clf.predict(test.reshape(1,-1))
    if response[0]==ytest[idx]:
        correct_results+=1
print("Accuracy: ",correct_results/test_size)        
```
